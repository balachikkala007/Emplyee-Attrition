```python
lr2=LogisticRegression()
lr2.fit(X_train1,y_train1)
print("Train Confusion Matrix:")
print(confusion_matrix(y_train1, lr2.predict(X_train1)))
print("Test Confusion Matrix:")
print(confusion_matrix(y_test1, lr2.predict(X_test1)))
print("Train Classification Report:")
print(classification_report(y_train1, lr2.predict(X_train1)))
print("Test Classification Report:")
print(classification_report(y_test1, lr2.predict(X_test1)))
```

```python
dt1=DecisionTreeClassifier()
dt1.fit(X_train1,y_train1)
print("Train Confusion Matrix:")
print(confusion_matrix(y_train1, dt1.predict(X_train1)))
print("Test Confusion Matrix:")
print(confusion_matrix(y_test1, dt1.predict(X_test1)))
print("Train Classification Report:")
print(classification_report(y_train1, dt1.predict(X_train1)))
print("Test Classification Report:")
print(classification_report(y_test1, dt1.predict(X_test1)))
```
